```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Creating a Figure And Axes
fig, ax = plt.subplots(figsize=(10,6))
# Setting the background color
fig.set_facecolor('lightgrey')
ax.set_facecolor('lightblue')
#plotting the data
x = [1,2,3,4,5,]
y = [23,45,24,12,23]
ax.plot(x,y,
        marker ='o',
        markersize = 5,
        markeredgecolor = 'black',
        markerfacecolor = 'red',
        linewidth = 2,
        linestyle = ':',
        color = 'green',
        alpha = 0.8)
ax.set_title('Line Plot', fontsize=10, color='black')
ax.set_xlabel('X-axis', fontsize=8, color='blue')
ax.set_ylabel('Y-axis', fontsize=8, color='blue')

ax.axhline(y=30, color='red', linestyle='--',linewidth=1)
ax.axvline(x=3, color='red',linestyle='--',linewidth=1)

ax.axhspan(30, 45, color='black', alpha=0.5)
ax.axvspan(0, 3, color='black', alpha=0.5)

ax.text(0.5, 42, 'Line Plot', fontsize=20, color='red')
```

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25], color='red', marker='o', linestyle='--', linewidth=2, markersize=8)
plt.plot([1,2,3,4,5], [2,4,6,8,10], color='blue', marker='x', linestyle='-', linewidth=2, markersize=8)
plt.xticks([1,2,3,4,5],['A','B','C','D','E'])
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
