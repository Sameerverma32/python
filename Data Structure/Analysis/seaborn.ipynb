```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
#Count plot
sns.countplot(df, x='survived', hue='sex')
plt.title('Survived vs Not Survived by gender')
plt.show()
```

```python
sns.countplot(df, x="class",palette='viridis')
plt.title('number of passenger by class')
plt.show()
```

```python
sns.countplot(df, y="class",palette='viridis')
plt.title('number of passenger by class')
plt.show()
```

```python
sns.barplot(df,x='class', y='survived', palette='coolwarm')
plt.title('survival rate by class')
plt.show()
```

```python
sns.histplot(df, x='age',bins=30, palette='coolwarm', kde=True)
plt.title('Survival rate by age')
plt.show()
```

```python
sns.violinplot(df, x='class', y='age', hue='sex', split=True, palette='Set2')
plt.title('Age distibution by class')
plt.show()
```

```python
sns.violinplot(df, x='class', y='age', hue='sex', split=True, palette='tab10')
plt.title('Age distibution by class')
plt.show()
```

```python
sns.violinplot(df, x='class', y='age', hue='sex', split=True, palette='terrain')
plt.title('Age distibution by class')
plt.show()
```

```python
sns.scatterplot(df, x='age', y='fare', hue='survived',palette='winter')
plt.title('Fare vs Age by survived')
plt.show
```

```python
sns.scatterplot(df, x='age', y='fare', hue='survived',palette='terrain')
plt.title('Fare vs Age by survived')
plt.show
```

```python
sns.pairplot(df[['age','fare','survived','pclass']],hue='survived',palette='terrain')
plt.show()
```

```python
sns.pairplot(df[['age','fare','survived','pclass']],hue='survived',palette='spring')
plt.show()
```
